```python
#Import libraries
import os, csv 
import re

import pandas as pd
```

```python
#Open the file "ca_sample_data_a.csv" in "read mode"

#We need to verify that the phone column is in the following format : ddd-ddd-dddd (where d is a digit) (i.e. 514-321-0909)
#We need to verify that the email column is in the following format : xxxxxxx@xxxxx.xxx (where x is an alphanumeric character)

records = []
bad_phone_recs = []
bad_email_recs = []
good_recs = []

regex_phone = '^(\\d{3}[-.]\\d{3}[-.]\\d{4})$'
regex_email = r'^[0-9a-zA-Z]([-.\w]*[0-9a-zA-Z_+])*@(([0-9a-zA-Z][-\w]*\.)+[a-zA-Z]{2,9})$'


filename = 'ca_sample_data_a.csv'
path = os.getcwd()
filepath = path + '\\sourcedata\\' + filename
outpath = path + '\\output\\'


#print(filepath)
fh = open (filepath, mode = 'r')
freader = csv.reader(fh)
next(freader,None)
for rec in freader:
    first_name,last_name,company_name,address,city,province,postal,phone1,email,web,exam_1,exam_2,exam_3 = rec
    #print(rec)
    records.append(rec)
fh.close()

for rec in records:
    #print(rec[7])
    #print(rec[8])
    
    found_phone = re.search(regex_phone,rec[7])
    found_email = re.search(regex_email,rec[8])
    
    if not found_phone:
        bad_phone_recs.append(rec)
        print('Phone is bad:',rec[7])
        
    if not found_email:
        bad_email_recs.append(rec)
        print('Email is bad:', rec[8])

    if found_phone and found_email:
        good_recs.append(rec)

#All records found with a phone that does not follow the given format are written to a file : bad_phone_recs.csv
#All records found with an email that does not follow the given format are written to a file : bad_email_recs.csv

#All good records should be written to a file : good_recs.csv

header = ['first_name','last_name','company_name','address','city','province','postal','phone1','email','web','exam_1','exam_2','exam_3']

fhw = open(outpath + 'bad_phone_recs.csv',mode='w', newline='')
fwriter = csv.writer(fhw)
fwriter.writerow(header)
fwriter.writerows(bad_phone_recs)
fhw.close()

fhw = open(outpath + 'bad_email_recs.csv',mode='w', newline='')
fwriter = csv.writer(fhw)
fwriter.writerow(header)
fwriter.writerows(bad_email_recs)
fhw.close()

fhw = open(outpath + 'good_recs.csv',mode='w', newline='')
fwriter = csv.writer(fhw)
fwriter.writerow(header)
fwriter.writerows(good_recs)
fhw.close()

    
```
